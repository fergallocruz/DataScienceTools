### Modificar los par치metros de una sola figura

```python

x = np.arange(0,5)
y = np.array([1,3,2,4,6])

fig = plt.figure(figsize=(3,3))
ax1 = fig.add_subplot(1,1,1)
ax1.plot(x,y)
```

```python

fig = plt.figure( figsize=(10,5))
grid = fig.add_gridspec(3,3)
ax1 = fig.add_subplot( grid[0,:])
ax2 = fig.add_subplot(grid[1:,0:2])
ax3 = fig.add_subplot(grid[1,2])
ax4 = fig.add_subplot(grid[2,2])

ax1.set_title('Grafica #1')
ax1.plot(np.arange(3,6), np.array([2,5,8]))
ax2.set_title('Grafica #2')
ax2.scatter(np.arange(1,5), np.arange(5,9))
ax3.set_title('Grafica #3')
ax4.set_title('Grafica #4')
# set the spacing between subplots
fig.tight_layout()
plt.show()
```

### Diferentes tipos de gr치ficas

```python
x = np.arange(-10,10,1)
y = x**2
mpl.rcParams['figure.figsize'] = (8,4)
plt.plot(x, y, color='#FF11FF', linestyle='-', linewidth=4, marker='o', markersize=12, markeredgecolor='purple', markeredgewidth=3)
```

### C칩mo personalizar 

```python
x = np.arange(-10,10,1)
y = x**2
mpl.rcParams['figure.figsize'] = (8,4)
mpl.rcParams['figure.facecolor'] = '#F2C53D'
mpl.rcParams['axes.facecolor'] = '#F2F0D5'
mpl.rcParams['axes.edgecolor'] = '#668C4A'
mpl.rcParams['axes.titlecolor'] = '#668C4A'
plt.plot(x, y, color='#A6BF4B', linestyle='-', linewidth=1, marker='o', markersize=8,markerfacecolor='#668C4A', markeredgecolor='#A6BF4B', markeredgewidth=3, )
plt.grid(True)
plt.xticks(x)
plt.yticks(np.arange(0,110,10));
```
