```python
import matplotlib.pyplot as plt
```

```python
x = [22,22,19,18,15]
plt.pie(x)
```

```python
from random import random

def rColor():
  return(random(), random(), random())
```

```python
x = [22,22,19,18,15]
labels = ['Jalisco', 'CDMX','Chiapas','Nuevo León','Yucatán']
sep_angle = [0,0,0,.2,0]
plt.pie(x, labels=labels, explode=sep_angle, shadow=True, autopct='%1.1f%%', colors=[rColor() for i in x])
plt.show()
```
