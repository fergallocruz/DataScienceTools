- Crea una lista con cada partido, donde los datos de cada partido sean un diccionario. Con las categorías como llave y el valor como el contenido.


```
# partidos
import csv
años = [2,3,4,5,6,7]
l = []
for i in años:
  f = open(f"atp_matches_201{i}.csv")
  data = csv.reader(f)
  headers = next(data)
  l+=list(data)

partidos = []

for p in l:
  partido = dict()
  for h, v in zip(headers, p):
    partido[h] = v
  partidos.append(partido)
partidos
```

# Filtrado de datos
- Haz un conjunto de todos los nombres de ganadores de partidos.


```
# ganadores

winners = {p['winner_name'] for p in partidos}
winners
```

- Haz un conjunto de todos los nombres de perdedores de partidos.


```
# perdedores
losers = {p['loser_name'] for p in partidos}
losers
```

- Enlista qué jugadores no han perdido ningún partido. 

```
# no han perdido
all_time_winners = winners - losers
all_time_winners
```

- ¿Cuántos jugadores no han perdido ningún partido?


```
# número de jugadores que no han perdido
len(all_time_winners)
```

- Enlista qué jugadores no han ganado ningún partido.

```
# no han ganado
all_time_losers = losers - winners
all_time_losers
```

- ¿Cuántos jugadores no han ganado ningún partido?

```
# número de jugadores que no han ganado
len(all_time_losers)
```

- Haz una lista diciendo cuántos partidos ha ganado cada jugador.

```
# cantidad de partidos ganados por jugador
wins_per_player = {}
for p in partidos:
  winner = p['winner_name']
  if(winner in list(wins_per_player.keys())):
    wins_per_player[winner] +=1
  else:
    wins_per_player[winner] = 1
wins_per_player
```

```
from collections import Counter
winners = [p['winner_name'] for p in partidos]
cuenta = Counter(winners)
ls_winners = [(jugador, cuenta[jugador]) for jugador in winners]
ls_winners
```

- Haz una lista diciendo cuántos partidos ha perdido cada jugador

```
# cantidad de partidos perdidos por jugador
lost_per_player = {}
for p in partidos:
  loser = p['loser_name']
  if(loser in list(lost_per_player.keys())):
    lost_per_player[loser] +=1
  else:
    lost_per_player[loser] =1
lost_per_player
```

- Obtén el jugador con más partidos ganados.

```
# jugador con más partidos ganados
max = 0
name = ''
for w in wins_per_player:
  if(wins_per_player[w]>max):
    name = w
    max = wins_per_player[w]
print(f'{name} has the top rank with {wins_per_player[name]} matches winned.')
```

- Obtén el jugador con más partidos perdidos

```
# jugador con más partidos perdidos
max = 0
name = ''
for l in lost_per_player:
  if(lost_per_player[l]>max):
    name = l
    max = lost_per_player[l]
print(f'{name} has lost the most matches with {lost_per_player[name]} matches lost.')
```

# Ordenamiento
- Obtén los 10 jugadores con más partidos ganados y especifica cuántos partidos ganaron.

```
# top 10
top_rank = sorted(wins_per_player.items(), key = lambda kv:(kv[1], kv[0]))
c=0
for t in reversed(top_rank[-10:]):
  c+=1
  print(f'{c}) {t[0]} has {t[1]} winned matches')
```

- Obtén los 10 jugadores con más partidos perdidos y especifica cuántos partidos perdieron.

```
# top 10
least_rank = sorted(lost_per_player.items(), key = lambda kv:(kv[1], kv[0]))
c=0
for t in reversed(least_rank[-10:]):
  c+=1
  print(f'{c}) {t[0]} has {t[1]} lost matches')
```

- Obtén una lista con los tipos de canchas que existen (pasto, arcilla, etc)

```
# Tipos de canchas
court_type = []
for p in partidos:
  if(p['surface'] not in court_type):
    court_type.append(p['surface'])
court_type
```

- Para cada tipo de cancha: Obtén una lista con los tres jugadores con más finales ganadas, y la cantidad de finales ganadas.

```
top_3_champions_per_court_type = dict()
for t in court_type:
  top_3_champions_per_court_type[t] = dict()

for p in partidos:
  if(p['winner_name'] in champions):
    if(p['winner_name'] in top_3_champions_per_court_type[p['surface']].keys()):
      top_3_champions_per_court_type[p['surface']][p['winner_name']]+=1
    else:
      top_3_champions_per_court_type[p['surface']][p['winner_name']] = 1
top_3_champions_per_court_type
max = 0
champion_of_champions = ''
for c in top_3_champions_per_court_type:
  print(c)
  top_3_per_court = sorted(top_3_champions_per_court_type[c].items(), key = lambda kv:(kv[1], kv[0]))
  c=0
  for t in reversed(top_3_per_court[-10:]):
    if (max < int(t[1])):
      max = int(t[1])
      champion_of_champions = t[0]
    c+=1
    print(f'{c}) {t[0]} has {t[1]} winned matches')
```
